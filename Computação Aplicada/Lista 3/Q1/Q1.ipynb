# Q1


Temos que a distribuição de carga volumétrica é dada por 

$$\displaystyle \rho{(r)} = \displaystyle\rho_0 \cdot \displaystyle e^{\displaystyle \frac{-r}{a}}$$ 

tal que , $(0 \leq r < \infty)$.

Vamos definir uma esfera tal como, $\theta$ sendo o ângulo azimutal e $\phi$ o ângulo zenital, dessa forma ficamos com. 

$$ dV = r^2 \ \sin{\phi} \ \ dr \ d\phi \ d\theta $$


## a)

Podemos escrever a carga $Q$ como: 
$$ Q = \displaystyle\iiint \left(\displaystyle\rho_0 \cdot \displaystyle e^{\displaystyle \frac{-r}{a}}\right) r^2 \ \sin{\phi} \ \ dr \ d\phi \ d\theta $$

Reduzindo chegamos em: 

$$ Q = \int^{2\pi}_0 \int^{\pi}_0 \int^{\infty}_0  \displaystyle\rho_0 \cdot \displaystyle e^{\left(\displaystyle \frac{-r}{a}\right) }r^2 \ \sin{\phi} \ \ dr \ d\phi \ d\theta $$

Vamos adotar para o raio de $]0,5]$ para a integral não ficar muito abstrata. Para plotagem do gráfico precisamos definir valores numéricos para nossas constantes, desta forma adotamos $\rho_0 = 3.2\cdot 10^{-4} \ C/m^3 \ e \ a= 1$ 

```python
import numpy as np
import sympy as sp
from math import *


r = sp.Symbol("r")
phi = sp.Symbol("phi")
theta = sp.Symbol("theta")
rho = 3.2e-4
a = 1
e0 = 8.85e-12
f =rho*sp.exp(-r/a)*r**2*sp.sin(phi)

Q = sp.integrate(sp.integrate(sp.integrate(f,(r,0,5)),(phi,0,sp.pi)),(theta,0,2*sp.pi))

Q.simplify()
```

## b)

Aplicando a lei de Gaus para campos elétricos temos que $$ \int_S \vec{E}\cdot d\vec{A} = \frac{q_{env}}{\epsilon_0}$$ 

logo temos que $$ E = \frac{Q}{4\pi \ \epsilon_0 \ r^2}$$ 

pensando agr que o campo eletríco cai com o inverso do quadrado do raio, faremos uma distribuição para o intervalo radial proposto.

```python
import matplotlib.pyplot as plt 

r = np.linspace(1,5,100)
E = lambda r: Q/(4*sp.pi*e0*r**2)
E(r)
plt.title("Comportamento do Campo Elétrico")
plt.xlabel("Raio da esfera")
plt.ylabel("Campo Elétrico")

plt.plot(r,E(r))
```
