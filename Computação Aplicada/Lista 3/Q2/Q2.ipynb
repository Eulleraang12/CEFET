No primeiro momento temos um circuito $RC$ simples, com o capacitor inicalmente descarregado e uma fonte $\varepsilon=V$ alimentando o circuito, logo aplicando lei de Krchhoff para malhas temos 

$$ -V + RI + V_c = 0 $$
$$ V = R \ \frac{dQ}{dt} + \frac{Q}{C} $$
resolvendo a EDO ficamos com a equação de carga de um capacitor

$$ Q_{(t)} = C\cdot V \left( 1 - e^{\displaystyle\frac{-t}{RC}} \right)$$

Vamos adaptar valores para termos um gráfico menos abstrado mostrando a carga e descarga no capacitor.

```python
q1 = lambda t: C*V*(1-sp.exp(-t/(R*C)))
plot(q1(t),(t,0,3*60), show = True, title="Carregamento do Capacitor",xlabel=" Tempo (s)", ylabel="Carga (C)")
```

Nesse segundo momento a chave é virada elimiando a fonte, e fazendo o capacitor descarregar na resintência.
Aplicando novamente malhas temos:

$$ RI + V_c = 0 $$
$$ R \ \frac{dQ}{dt} = - \frac{Q}{C} $$

integrando nossa EDO homogênia ficamos com 
$$ Q_{(t)} = Q_0 \ e^{\displaystyle\frac{-t}{RC}}  $$

onde $Q_0$ é a carga inicial que foi carregada.

```python
q0 = 25e-3 #Simulando que foi carregado como o de cima Carga de 25e-3C

q2 = lambda t: q0*sp.exp(-t/(R*C))
plot(q2(t),(t,0,3*60), show = True, title="Descarregamento do Capacitor",xlabel=" Tempo (s)", ylabel="Carga (C)")
```

```python
p = plot(q1(t),q2(t),(t,0,3*60),show=False,title="Carga e Descarga do Capacitor",xlabel=" Tempo (s)", ylabel="Carga (C)")
p[1].line_color = 'r'
p.show()
```

Agora faremos uma análise durante todo o tempo com o ato chavear o circuito $RC$ começando com ele chaveado a fonte $AC$.
